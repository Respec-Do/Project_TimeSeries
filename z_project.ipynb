```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

def get_vif(features):
    vif = pd.DataFrame()
    vif['vif_score'] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
    vif['feature'] = features.columns
    return vif
```

VIF를 통해 Google과 GLD 간의 다중공산성 문제를 확인합니다.  
확인한 결과, 두 vif_score가 모두 5를 넘지 않는 것으로 보아 Google과 GLD간에는 서로 독립적이라고 판단할 수 있었습니다.

```python
rate_f_df = rate_f_df.dropna()
get_vif(rate_f_df)
```

cumsum()을 이용하여, 일간 수익률을 확인합니다.  
확인했을 때, GLD는 2010년에 비해 일간 수익률의 변동이 크게 나타나지 않는 반면,   
Google은 2010년과 비교하면 약 7배 정도 증가한 것을 확인할 수 있었습니다. 

```python
# 각 원소들의 누적합 : cumsum()
# 일간 수익률

rate_f_df.cumsum().apply(np.exp).plot(figsize=(12, 6))
plt.show()
```

월간 수익률을 확인합니다.  
일간 수익률에서, resample을 사용하여 데이터를 월 단위로 다시 샘플링합니다.  
이때, last()를 사용하여 월말을 기준으로 시각화합니다.

```python
# 월간 수익률
rate_f_df.cumsum().apply(np.exp).resample('1m').last().plot(figsize=(12, 6))
plt.show()
```

### Golden Cross(골드 크로스) & Death Cross(데드 크로스)

- 골드 크로스와 데드 크로스는 시장에서 기술적 분석의 주요 개념으로 사용되며, 이동 평균(MA)를 통해 매수와 매도 신호를 파악하는데 큰 도움을 줍니다.

1. 골드 크로스

- 골드 크로스는 단기 이동 평균선이 장기 이동 평균선을 아래에서 위로 교차할 때 발생합니다. 이는 주가가 상승 추세로 전환할 가능성이 높다는 신호로 해석됩니다.
- 골드 크로스는 일반적으로 강세 시장의 시작을 나타내며, 투자자들에게 매수 신호로 간주됩니다.
- 이 교차는 시장 심리가 긍정적으로 변화하고 있음을 의미합니다.


2. 데드 크로스

- 데드 크로스는 단기 이동 평균선이 장기 이동 평균선을 위에서 아래로 교차할 때 발생합니다. 이는 주가가 하락 추세로 전환할 가능성이 높다는 신호로 해석됩니다.
- 일반적으로 약세 시장의 시작을 나타내며, 투자자들에게 매도 신호로 간주됩니다.
- 이 교차는 시장 심리가 부정적으로 변화하고 있음을 의미합니다.

`골드크로스`와 `데드크로스`를 통해 구글 수익률 분석을 진행합니다.

시각화하여 확인한 결과, 차분한 MR에서 값이 다소 분산되어있지만 평균은 일정한 편으로 확인이 되었습니다.

ndiffs를 활용하여 가장 적합한 차분계수를 찾습니다.

```python
from pmdarima.arima import ndiffs
# KPSS(Kwaiatkowski-Phillips-Schmidt-Shin)
# 차분을 진행하는 것이 필요할 지 결정하기 위해 사용하는 한 가지 검정 방법
# 영가설(귀무가설)을 "데이터에 정상성이 나타난다."로 설정한 뒤
# 영가설이 거짓이라는 증거를 찾는 알고리즘이다.
kpss_diffs = ndiffs(y_train, alpha=0.05, test='kpss', max_d=6)
adf_diffs = ndiffs(y_train, alpha=0.05, test='adf', max_d=6)
pp_diffs = ndiffs(y_train, alpha=0.05, test='pp', max_d=6)

n_diffs = max(kpss_diffs, adf_diffs, pp_diffs)

print(f'd = {n_diffs}')
```

- acf와 pacf를 시각화하여 확인했을 때 판단했던 차분계수는 1이었습니다.
- ndiffs를 통해 가장 적합한 차분계수를 나타낸 결과, 1로 나타났습니다.
- 이를 통해 차분계수는 1로 두고 훈련을 진행합니다.

```python
import pmdarima as pm

model = pm.auto_arima(y=y_train, 
                      d=1, 
                      start_p=0, max_p=10, 
                      start_q=0, max_q=10, 
                      m=1, seasonal=False, 
                      stepwise=True, 
                      trace=True)
```

```python
model.fit(y_train)
```

```python
# Prob(Q), 융-박스 검정 통계량
# 영가설: 잔차가 백색잡음 시계열을 따른다.
# 0.05 이상: 서로 독립이고 동일한 분포를 따른다.

# Prob(H), 이분산성 검정 통계량 (잔차의 분산이 일정한지 검사)
# 영가설: 잔차가 이분산성을 띄지 않는다.
# 0.05 이상 : 잔차의 분산이 일정하다.

# Prob(JB), 자크-베라 검정 통계량
# 영가설: 잔차가 정규성을 따른다.
# 0.05 이상: 일정한 평균과 분산을 따른다.

# Skew(쏠린 정도, 왜도)
# 0에 가까워야 한다.

# Kurtosis(뾰족한 정도, 첨도)
# 3에 가까워야 한다.
print(model.summary())
```

```python
train_df = deep_google_df.iloc[:int(len(deep_google_df) * 0.8)]
train_df
```

```python
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics
import itertools

# changepoint_prior_scale : trend의 변화하는 크기를 반영하는 정도이다. 0.06가 default
# seasonality_prior_scale : 계절성을 반영하는 단위이다.
# seasonality_mode : 계절성으로 나타나는 효과를 더해 나갈지, 곱해나갈지 정한다.
search_space = {
    'changepoint_prior_scale': [0.05, 0.1, 0.5, 1.0, 5.0, 10.0],
    'seasonality_prior_scale': [0.05, 0.1, 1.0, 10.0],
    'seasonality_mode': ['additive', 'multiplicative']
}

# itertools.product(): 각 요소들의 모든 경우의 수 조합으로 생성

param_combined = [dict(zip(search_space.keys(), v)) for v in itertools.product(*search_space.values())]

train_len = int(len(deep_google_df) * 0.8)
test_len = int(len(deep_google_df) * 0.2)

train_size = f'{train_len} days'
test_size = f'{test_len} days'
train_df = deep_google_df.iloc[: train_len]
test_df = deep_google_df.iloc[train_len: ]

mapes = []
for param in param_combined:
    model = Prophet(**param)
    model.fit(train_df)

    # 'threads' 옵션은 메모리 사용량은 낮지만 CPU 바운드 작업에는 효과적이지 않을 수 있다.
    # 'dask' 옵션은 대규모의 데이터를 처리하는 데 효과적이다.
    # 'processes' 옵션은 각각의 작업을 별도의 프로세스로 실행하기 때문에 CPU 바운드 작업에 효과적이지만,
    # 메모리 사용량이 높을 수 있다.
    cv_df = cross_validation(model, initial=train_size, period='20 days', horizon=test_size, parallel='processes')
    df_p = performance_metrics(cv_df, rolling_window=1)
    mapes.append(df_p['mape'].values[0])

tuning_result = pd.DataFrame(param_combined)
tuning_result['mape'] = mapes
```

```python
tuning_result.sort_values(by='mape')
```

- 점수가 가장 잘 나온 parameter를 입력하여 다시 모델을 정의합니다.
- 이때, 계절성이 없는 시계열인 경우, additive를 사용합니다.
- additive 모델은 계절성의 변동이 일정하게 발생한다고 가정하며, 예측 시 계절성의 변동을 추가적으로 더합니다.
- multiplicative를 사용할 경우, 계절성의 변동이 주기적으로 변화한다고 가정하며, 예측 시 계절성의 변동을 곱합니다.
- 이에 multiplicative를 계절성이 없는 시계열데이터에 사용할 경우, 계절성의 변동을 같이 나타내게 되어 예측의 변동성이 커져 오히려 모델의 예측성능이 저하됩니다.

```python
model = Prophet(changepoint_prior_scale=1.0, 
                seasonality_prior_scale=0.05, 
                seasonality_mode='additive')

model.fit(deep_google_df)
future = model.make_future_dataframe(periods=365)
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']][221:]
```

- 이에, multiplicative를 계절성이 없는 시계열데이터에 사용할 경우, 계절성의 변동을 같이 나타내게 되어 예측의 변동성이 커져 오히려 모델의 예측성능이 저하됩니다.

#### 🚩결론
